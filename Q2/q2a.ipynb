```python
import pandas as pd
def convt_to_dict(txt_file):
     """ Taking input as text file and converting it into a Dictionary with keys as Version,Definition and Sequence"""
     with open(txt_file) as file:
        info = file.read()
        
     info = info.split(">")
     while True:
        try:
            info.remove("")
        except:
            break 
            
     nucleotide_list, nucleotides = [], {"Version": [], "Defination": [], "Sequence": []}
     for i in info:
        temp, seq = i.split("\n", maxsplit=1)
        nucleotides['Version'].append(temp.split(" ", maxsplit=1)[0])
        nucleotides['Defination'].append(temp.split(" ", maxsplit=1)[1])
        nucleotides["Sequence"].append("".join(seq.split("\n")))
     return nucleotides
```

```python
# Storing the dictionary as "sequences"
sequences = convt_to_dict("./../ProteinAlign.txt")
len(sequences['Sequence'][0])
```

```python
"""
    Import the scoring scheme "BLOSUM62.txt", to score the 
    simliarities between the various protiens to be used in
    making the Phylogenic Tree.
"""

with open("./../BLOSUM62.txt") as file:
    info = file.read()
info = info.split("\n")
info = [ x.split() for x in info ]
scores = pd.DataFrame(info)
temp = list(scores.iloc[0])
temp[0] = "Remove"
scores.columns, scores.index = temp, temp
scores = scores.drop(scores.index[0])
scores = scores.drop(scores.columns[0], axis=1)
scores.head()
```

```python
def get_score(str1, str2):
    """
        Used to find the degree of difference between 2 given strings
        based on the data from "./../BLOSUM62.txt".
    """
    global scores
    m = len(str2)
    str1 = "*".join( str1.split("-") )
    str2 = "*".join( str2.split("-") )
    score = 0
    for i in range(m):
        score += int(scores.at[str1[i], str2[i]])
    return score
```

```python
distance_matrix = []
for i in range(len(sequences['Sequence'])):
    distance = []
    for j in range(len(sequences['Sequence'])):
        distance.append( get_score(sequences['Sequence'][i], sequences['Sequence'][j]))
    distance_matrix.append(distance)
```

```python
df = pd.DataFrame(distance_matrix)
names = {0:"A", 1:"B", 2:"C", 3:"D", 4:"E",
         5: "F", 6:"G", 7:"H", 8:"I", 9:"J"}
df.rename(names, inplace=True)
df.columns = list(names.values())
df
```

```python
df.to_csv("Pdistance.txt")
```
