```python
import pandas as pd
```

```python
distance_data = pd.read_csv("./../Ndistance.txt", index_col=0)
names = {0:"A", 1:"B", 2:"C", 3:"D", 4:"E", 5: "F", 6:"G", 7:"H", 8:"I", 9:"J"}
distance_data.rename(index=names, inplace=True)
distance_data.columns = list(names.values())
distance_data
```

```python
half_list = []
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
distance_data = distance_data.to_numpy()
for i in range(10):
    row = []
    for j in range(i):
        row.append(distance_data[i][j])
    half_list.append(row)
del row
```

```python
def lowest_value(table):
    """ Given an 2d array the function returns the lowest value in the array."""
    min_cell = float("inf")
    x, y = -1, -1

    for i in range(len(table)):
        for j in range(len(table[i])):
            if table[i][j] < min_cell:
                min_cell = table[i][j]
                x, y = i, j

    return x, y

def join_labels(labels, a, b):
    """ Used for keeping track of combined nodes easier.
        Deletes the larger valued node after the combination """
    
    if b < a:
        a, b = b, a
    labels[a] = "(" + labels[a] + "," + labels[b] + ")"
    del labels[b]

def join_table(table, a, b):
    """
        Given the coordinates of the smallest element, this function joins 
        the corresponding rows according to the UPGMA algorithm. 
        """
    if b < a:
        a, b = b, a

    row = []
    for i in range(0, a):
        row.append((table[a][i] + table[b][i])/2)
    table[a] = row
    
    for i in range(a+1, b):
        table[i][a] = (table[i][a]+table[b][i])/2
        
    for i in range(b+1, len(table)):
        table[i][a] = (table[i][a]+table[i][b])/2
        del table[i][b]

    del table[b]
```

```python
def UPGMA(table, labels):
    """
        Phylogenetic tree in The Newick tree format """
    while len(labels) > 1:
        x, y = lowest_value(table)
        join_table(table, x, y)
        join_labels(labels, x, y)
    return labels[0]
```

```python
UPGMA(half_list, labels)
```
