```python
def convt_to_dict(txt_file):
    """ Function that Takes text file as an input and converting it into a Dictionary with keys as 
    Version, Definition and Sequence"""
     with open(txt_file) as file:
        info = file.read()
        
     info = info.split(">")
     while True:
        try:
            info.remove("")
        except:
            break 
            
     nucleotide_list, nucleotides = [], {"Version": [], "Definition": [], "Sequence": []}
     for i in info:
        temp, seq = i.split("\n", maxsplit=1)
        nucleotides['Version'].append(temp.split(" ", maxsplit=1)[0])
        nucleotides['Defination'].append(temp.split(" ", maxsplit=1)[1])
        nucleotides["Sequence"].append("".join(seq.split("\n")))
     return nucleotides
```

```python
# Storing the dictionary as "sequences"
sequences = convt_to_dict("./../NucleotideAlign.txt")
length= len(sequences['Sequence'][0])
print(length)
```

```python
# Calculating the distance between seq1 and seq2= Mismatches/total length 
def distance(seq1,seq2):
    slen=len(seq1)
    notmatch=0
    dash=0
    
    for i in range(slen):
        if seq1[i]==seq2[i]:
            if seq1[i]=='-':
                dash+=1
        else:
            notmatch+=1
    return (notmatch/(slen-dash))
```

```python
#Creating the n*n distance matrix by iterating over the given n sequences
dist_matrix=[]
seqlen=len(sequences['Sequence'])
for i in range(seqlen):
    dist=[]
    for j in range(seqlen):
         dist.append(distance(sequences['Sequence'][i], sequences['Sequence'][j]))
    dist_matrix.append(dist)
```

```python
import pandas as pd
df = pd.DataFrame(dist_matrix)
df
```

```python
# Converting the matrix into csv 
df.to_csv("./Ndistance.txt")
```
